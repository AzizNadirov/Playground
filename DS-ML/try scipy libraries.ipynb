```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

$f(x) = x^2+x+1$

```python
def f(x):
    return x **2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize = 17)
plt.ylabel("f(x)", fontsize = 17)
plt.plot(x_1, f(x_1))
plt.show()
```

```python
def df(x):
    return 2*x+1 
```

```python
plt.figure(figsize = [15, 5])
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.title("cost function", fontsize = 17)
plt.xlabel("X", fontsize = 17)
plt.ylabel("f(x)", fontsize = 17)
plt.plot(x_1, f(x_1), color = "blue", linewidth = 3)

plt.subplot(1,2,2)
plt.xlim(-2,6)
plt.ylim(-3, 6)
plt.title("slope of cost function", fontsize = 17)
plt.xlabel("X", fontsize = 17)
plt.ylabel("df(x)", fontsize = 17)
plt.plot(x_1, df(x_1), color = "blue", linewidth = 3)
```
